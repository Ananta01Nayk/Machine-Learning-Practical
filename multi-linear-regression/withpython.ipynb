```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.datasets import load_diabetes
```

```python
x,y = load_diabetes(return_X_y=True)
```

```python
class MYReg:
    def __init__(self):
        pass
    def fit(self,x_train,y_train):
        """for the fit the model we need to find the required ment for bitas you know the formula is like this B=(x.t*x)-1 xy.t"""
        self.x_train = x_train
        self.y_train = y_train
        self.x_train = np.insert(self.x_train,0,1,axis=1)
        x_train_E = np.dot(self.x_train.T,self.x_train)
        first_part = np.linalg.inv(x_train_E)
        
        second_part = np.dot(self.x_train.T,self.y_train)
        betas = np.dot(first_part,second_part)
        self.intercept_ = betas[0]
        self.coef_ = betas[1:]
    
    def predict(self,x_test):
        self.result = np.dot(x_test,self.coef_) + self.intercept_
        
        return self.result
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
np.insert(x_train,0,1,axis=1).shape
```

```python
mylr = MYReg()
```

```python
y_train.shape
```

```python
mylr.fit(x_train,y_train)
```

```python
y_pred=mylr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
