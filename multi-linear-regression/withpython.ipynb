```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
class MYReg:
    def __init__(self):
        pass
    def fit(self,x_train,y_train):
        """for the fit the model we need to find the required ment for bitas you know the formula is like this B=(x.t*x)-1 xy.t"""
        self.x_train = np.insert(x_train,0,1,axis=0)
        self.y_train = y_train
        x_train_E = np.dot(self.x_train.T,self.x_train)
        first_part = np.linalg.inv(x_train_E)
        second_part = np.dot(self.x_train,self.y_train.T)
        biteas = np.dot(first_part,second_part)

        return biteas
```

```python
x_train,x_test,y_train,y_test = train_test_split(df['cgpa'],df['package'],test_size=0.2,random_state=43)
```

```python
mylr = MYReg()
```

```python
mylr.fit(x_train,x_test)
```
