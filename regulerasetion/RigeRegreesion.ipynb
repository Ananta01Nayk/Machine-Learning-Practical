```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ridge_regression
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
```

```python
x,y = load_diabetes(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
class MyRige:
    def __init__(self,alpha=0.1):
        self.alpha=alpha
    def fit(self,x_train,y_train):
        """formula is y=x*cofie_ + intercept_
       betas =(x^-T*X+alpha*I)X^-y*X """
        x_train = np.insert(x_train,0,1,axis=1)
        I= np.eye(len(x_train[0]))

        metrixs_invers_dot = np.dot(x_train.T,x_train)
        metrix_invers = np.linalg.inv(metrixs_invers_dot + self.alpha*I) 
        metrix_not_invers = np.dot(x_train.T,y_train)

        betas = np.dot(metrix_invers,metrix_not_invers)
        self.intercept_ = betas[0]
        self.cofie_ = betas[1:]

    def predic(self,x_test):

        return np.dot(x_test,self.cofie_) + self.intercept_
```

```python
my_Reg = MyRige()
```

```python
my_Reg.fit(x_train,y_train)
```

```python
y_pred=my_Reg.predic(x_test)
```

```python
r2_score(y_test,y_pred)
```
