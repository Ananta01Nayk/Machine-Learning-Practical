```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
def step(z):
    if z>0:
        return 1
    if z<0:
        return 0
```

```python
def perceptron(X,Y):
    x= np.insert(X,0,1,axis=1)
    widths = np.ones(x.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,100)
        z = np.dot(x[j],widths)
        y_pred = step(z)
        widths = widths + lr*((Y[j]-y_pred)*x[j])
    return widths[0],widths[1:]
```

```python
intercept_,cofie_=perceptron(x,y)
m = (cofie_[0]/cofie_[1])
b = (intercept_/cofie_[1])
```
