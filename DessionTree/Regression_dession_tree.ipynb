```python
#import importent library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
#make data set
x = np.random.rand(200,1)-5
y= x**2-2*np.random.rand(200,1)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#create model
tree_r = DecisionTreeRegressor(max_depth=2)
tree_r.fit(x,y)
```

```python
#tree regresor
from sklearn.tree import export_text
tree = export_text(tree_r)
print(tree)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))  # Customize size
plot_tree(tree_r,filled=True, rounded=True)
plt.show()
```
