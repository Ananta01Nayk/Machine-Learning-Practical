```python
#required library
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
```

```python
x,y = load_breast_cancer(return_X_y=True)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def perceptron(x,y):
    x = np.insert(x,0,1,axis=1)
    weigths = np.ones(x.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,100)
        z = np.dot(x[j],weigths)
        #call the sigmoid function to calculate z
        y_pred = sigmoid(z)

        #update the new weights
        weigths = weigths + lr*(y[j]-y_pred)*x[j]

    return weigths[0],weigths[1:]
```

```python
intercept_,cofe_ = perceptron(x,y)
```

```python
m_my = cofe_[0]/cofe_[1]
b_my = intercept_/cofe_[1]
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_[0][0]
```

```python
m_log = lr.coef_[0][0]/lr.coef_[0][1]
b_log = lr.intercept_/lr.coef_[0][1]
```

```python
m_log
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_input = np.linspace(x[0].mean(), x[1].mean(), len(x))
y_input_my =  m_my * x_input + b_my
y_input_log = m_log * x_input + b_log
plt.plot(x_input,y_input_my)
plt.plot(x_input,y_input_log)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = mx + b")
plt.grid(True)
plt.show()
```
